```python
# create one-number metrics for each player
grouped_data = df.groupby(['player_name', 'team'])

# calculate proportions for each evaluation_code
summary = grouped_data.agg(
    Perfect = ('evaluation_code', lambda x: (x == '#').mean()),
    Good = ('evaluation_code', lambda x: (x == '+').mean()),
    OK = ('evaluation_code', lambda x: (x == '!').mean()),
    PoorA = ('evaluation_code', lambda x: (x == '/').mean()),
    PoorB = ('evaluation_code', lambda x: (x == '-').mean()),
    Error = ('evaluation_code', lambda x: (x == '=').mean()),
    count = ('evaluation_code', 'count')
).reset_index()

# add new columns 'dig_score' and 'in_system'
summary['dig_score'] = 3*summary['Perfect'] + 2*summary['Good'] + 1.5*summary['OK'] + summary['PoorA'] + summary['PoorB']
summary['in_system'] = summary['Perfect'] + summary['Good']

# filter rows where 'count' is greater than or equal to 50
result = summary[summary['count'] >= 50]
```

```python
# verify equal length data sets
len(result) == len(df_hp)
```

```python
result.sort_values(by = 'dig_score', ascending = False).head()
```

```python
# left join the height and position data onto the defensive scores data
scores = result.merge(df_hp, left_on = 'player_name', right_on = 'Player')
scores
```
